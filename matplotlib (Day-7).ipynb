```python
x = np.arange(0,16,2)
plt.plot(x,[x1**2 for x1 in x])
plt.show()
```

```python
x = np.arange(0,16,2)
plt.plot(x,[x1 for x1 in x])
plt.plot(x,[x1**2 for x1 in x])
plt.plot(x,[x1**3 for x1 in x])
plt.show()
```

```python
x = np.arange(0,100,2)
plt.plot(x,[x1 for x1 in x])
plt.plot(x,[x1**2 for x1 in x])
plt.plot(x,[x1**3 for x1 in x])
plt.axis([0,20,0,80])
plt.show()
```

```python
x = [10,20,30,40]
y = [20,30,20,45]

x1 = [11,22,33,44]
y1 = [22,34,56,21]
plt.plot(x,y,'g',label ='sales',linewidth=10)
plt.plot(x1,y1,'y',label='profit',linewidth=5)
plt.title('first graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True, colour='cyan')
plt.show()
```

```python
x = ['10','12','14','16','18']
y = [15,10,12,18,20]
plt.plot(x,y)
for i in range(len(x)):
    print('i=',i,'and y[i]=',y[i])
    plt.text(i,y[i],y[i],ha='center',va='bottom')
```
